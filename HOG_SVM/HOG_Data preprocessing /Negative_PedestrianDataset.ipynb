```python
# Load pedestrian dataset
dir_dataset = os.path.join('/content/drive', 'MyDrive/Đồ án CV/PennFudanPed')
dir_images = os.path.join(dir_dataset, 'PNGImages')
```

```python
pedestrian_images = os.listdir(dir_images)
```

```python
# Pre-trained pedestrian detection model
hog = cv2.HOGDescriptor()
hog.setSVMDetector(cv2.HOGDescriptor_getDefaultPeopleDetector())
```

```python
# Sliding window parameters
window_size = (64, 128)  # Size of the sliding window
stride = 16  # Stride for window movement
```

```python
background_samples = []  # List to store background samples
regions = []
```

```python
# Iterate over pedestrian images
for image_name in pedestrian_images:
    image_path = dir_images + '/' + image_name

    # Load image
    image = cv2.imread(image_path)

    # Detect pedestrians in the image
    boxes, _ = hog.detectMultiScale(image)

    pedestrian_regions = []
    # Extract pedestrian regions
    for (x, y, w, h) in boxes:
      pedestrian_regions.append(image[y:y+h, x:x+w])
      # Iterate over pedestrian regions
      for region in pedestrian_regions:
      # Display the pedestrian region on the image
        cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 0), 2)  # Draw a green bounding box

    cv2_imshow(image)


    # Sliding window for background extraction
    for y in range(0, image.shape[0] - window_size[1], stride):
        for x in range(0, image.shape[1] - window_size[0], stride):
            # Check if the window overlaps with any pedestrian bounding box
            is_overlapping = False
            for (bx, by, bw, bh) in boxes:
                if x < bx + bw and x + window_size[0] > bx and y < by + bh and y + window_size[1] > by:
                    is_overlapping = True
                    break

            # If the window does not overlap with any pedestrian, consider it as a background sample
            if not is_overlapping:
                background_samples.append(image[y:y+window_size[1], x:x+window_size[0]])
```

```python
# Display or save the background samples as needed
for i, sample in enumerate(background_samples):
    cv2_imshow(sample)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    # Save the background sample to a file
    cv2.imwrite(f'background_sample_{i}.jpg', sample)
```
